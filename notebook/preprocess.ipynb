### **Customer Feedback Dataset**

```python
from textblob import TextBlob
```

```python
customer_feedback.head()
```

```python
def translate_to_english(text):
    try:
        if isinstance(text, str):
            return str(TextBlob(text).translate(from_lang='bg', to='en'))
        return text
    except Exception as e:
        return text
```

```python
if 'Review' in customer_feedback.columns:
    customer_feedback['translated_text'] = customer_feedback['Review'].apply(translate_to_english)
```

```python
customer_feedback.head()
```

```python
customer_feedback['cleaned_text'] = customer_feedback['translated_text'].apply(clean_text)
```

```python
customer_feedback['reviews'] = customer_feedback['translated_text']
```
