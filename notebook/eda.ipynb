### **Merged Dataset**

```python
merged_data = pd.read_csv('../data/clean/merged-labeled/merged_labeled_dataset.csv')
```

```python
merged_data.info()
```

```python
merged_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
merged_data_info = merged_data.info()

# Show the distribution of sentiment labels
sentiment_distribution = merged_data['Sentiment'].value_counts()

# Plot the sentiment distribution
plt.figure(figsize=(8, 5))
sentiment_distribution.plot(kind='bar', title='Sentiment Distribution', xlabel='Sentiment', ylabel='Count')
plt.show()
```

```python
merged_data['review_length'] = merged_data['cleaned_text'].apply(lambda x: len(str(x).split()))
```

```python
plt.figure(figsize=(8, 5))
plt.hist(merged_data['review_length'], bins=20, edgecolor='k')
plt.title('Review Length Distribution')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

```python
review_length_stats = merged_data['review_length'].describe()
```
