```python
import pandas as pd
```

```python
amazon = pd.read_csv('../data/clean/unlabeled/amazon_preprocessed.csv')
amazon_reviews = pd.read_csv('../data/clean/unlabeled/amazon_reviews_preprocessed.csv')
customer_feedback = pd.read_csv('../data/clean/unlabeled/Customer_feedback_preprocessed.csv')
redmi = pd.read_csv('../data/clean/unlabeled/redmi6_preprocessed.csv', encoding='ISO-8859-1')
sentiment_analysis = pd.read_csv('../data/clean/unlabeled/sentiment_analysis_preprocessed.csv')
```

```python
def map_sentiment(rating):
    if rating >= 4.0:
        return 'Positive'
    elif 3.0 <= rating < 4.0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
amazon['rating'] = pd.to_numeric(amazon['rating'], errors='coerce')
```

```python
amazon['rating'].isnull().sum()
amazon = amazon.dropna(subset=['rating'])
```

```python
amazon['Sentiment'] = amazon['rating'].apply(map_sentiment)
```

```python
amazon.head()
```

```python
amazon_labeled = amazon[['cleaned_text', 'Sentiment']]
amazon_labeled.to_csv('../data/clean/labeled/amazon_labeled.csv', index=False)
```

```python
amazon_reviews.head()
```

```python
amazon_reviews['overall'] = pd.to_numeric(amazon_reviews['overall'], errors='coerce')
```

```python
amazon_reviews['Sentiment'] = amazon_reviews['overall'].apply(map_sentiment)
```

```python
amazon_reviews.head()
```

```python
amazon_labeled = amazon_reviews[['cleaned_text', 'Sentiment']]
amazon_labeled.to_csv('../data/clean/labeled/amazon_reviews_labeled.csv', index=False)
```

```python
print(customer_feedback['Category'].unique())
```

```python
category_to_sentiment = {
    'compliments': 'Positive',
    'complaints': 'Negative',
    'mixed': 'Neutral',
    'suggestions': 'Neutral'
}
```

```python
customer_feedback['Sentiment'] = customer_feedback['Category'].map(category_to_sentiment)
```

```python
customer_labeled = customer_feedback[['cleaned_text', 'Sentiment']]
customer_labeled.to_csv('../data/clean/labeled/customer_feedback_labeled.csv', index=False)
```

```python
redmi.head()
```

```python
redmi['numeric_rating'] = redmi['Rating'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
redmi['Sentiment'] = redmi['numeric_rating'].apply(map_sentiment)
```

```python
redmi.head()
```

```python
redmi_labeled = redmi[['cleaned_text', 'Sentiment']]
redmi_labeled.to_csv('../data/clean/labeled/redmi6_labeled.csv', index=False)
```

### **Merge All Dataset**

```python
amazon_l = pd.read_csv('../data/clean/labeled/amazon_labeled.csv')
amazon_reviews_l = pd.read_csv('../data/clean/labeled/amazon_reviews_labeled.csv')
customer_l = pd.read_csv('../data/clean/labeled/customer_feedback_labeled.csv')
redmi_l = pd.read_csv('../data/clean/labeled/redmi6_labeled.csv')
sentiment_l = pd.read_csv('../data/clean/labeled/sentiment_analysis_labeled.csv')
```

```python
merged_data = pd.concat([amazon_l, amazon_reviews_l, customer_l, redmi_l, sentiment_l], ignore_index=True)
```

```python
merged_data.head()
```

```python
merged_data['Sentiment'] = merged_data['Sentiment'].str.strip().str.lower()
merged_data['Sentiment'] = merged_data['Sentiment'].replace({
    'negative': 'Negative',
    'positive': 'Positive',
    'neutral': 'Neutral'
})
```

```python
merged_data.isnull().sum()
```

```python
merged_data.dropna(subset=['cleaned_text'], inplace=True)
```

```python
merged_data.isnull().sum()
```

```python
merged_data_path = '../data/clean/merged-labeled/merged_labeled_dataset.csv'
merged_data.to_csv(merged_data_path, index=False)
```
